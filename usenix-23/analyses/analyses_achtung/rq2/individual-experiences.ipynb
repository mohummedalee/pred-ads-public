```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

DATA_DIR = '../data/'
```

```python
from collections import defaultdict

def count_code_props(codes, total_count=None, norm=True):
    # takes dict from adid -> codes and returns dict of code proportions if norm=True, just frequencies if False
    counts = defaultdict(lambda: 0)
    for aid in codes:
        # in case of multiple codes, count each one -- essentially computing fraction of codes and not ads here
        for code in codes[aid].split(';'):
            if "Can't determine" not in code and code != 'UNCAT':
                counts[code] += 1
                
    if norm:
        if total_count is not None:
            return {c: counts[c]/total_count for c in counts}
        else:
            raise("participant total ad count (total_count) not provided for denominator")
    else:
        return counts
```

### Analysis on how individual experiences vary:
- Ali: What fraction of each participant's ads are clickbait or scam or prohibited?
- Elissa: what fraction of ads does each participant dislike? This is an alternate analysis from what fraction of ads is clickbait or prohibited or some normative judgment that we pass
- Elissa: what fraction of ads does a participant dislike for X vs. Y vs. Z reasons? This can only happen if we have well-informed groups of reasons of dislike.

```python
codes = pd.read_csv(os.path.join(DATA_DIR, 'ad_codes.tsv'), sep='\t')
print('codes shape:', codes.shape)
```

```python
# load PID to ADID mapping of annotated ads
pid_adid_files = [os.path.join(DATA_DIR, f'surveys/survey-{n}/survey{n}_coding_subsample.tsv') for n in range(1, 8)]
pid_adid_files.append(os.path.join(DATA_DIR, 'surveys/survey-1/survey1_batch2_coding_subsample.tsv'))

pid_adid = pd.read_csv(pid_adid_files[0], sep='\t')
for f in pid_adid_files[1:]:
    pid_adid = pd.concat([pid_adid, pd.read_csv(f, sep='\t')], ignore_index=True)
    
active_pids = open(os.path.join(DATA_DIR, 'pid_active_contribs.txt'), 'r').readlines()
active_pids = [p.strip() for p in active_pids]

# only keep PID ADID mapping for active participants
pid_adid = pid_adid[pid_adid['pid'].isin(active_pids)]
print('pid_adid shape:', pid_adid.shape)   
```

```python
# number of per-participant annotated ads
part_coded_counts = {}
# fraction of per-code makeup in participant diet, out of all coded ads
part_code_props = {}

for pid in active_pids:
    pid_ads = pid_adid[pid_adid['pid'] == pid]
    pid_coded_ads = pd.merge(pid_ads, codes, how='inner', left_on='adid', right_on='adid')

    part_coded_counts[pid] = len(pid_coded_ads['adid'].unique())
    
    code_props = count_code_props(dict(pid_coded_ads[['adid', 'codes']].values), part_coded_counts[pid], norm=True)
    part_code_props[pid] = code_props
```
