## Forms of Harm

For now, throw away Manjot's codes.
Two broad forms of harms:
- global: harmful + clickbait. Bad for everyone
- contextual: sensitive. Bad in certain contexts

For both of these, we need a better sense of the demographics of people consuming the top 20 percentile of the harm.

Consider evaluating fairness metrics such as TPR bias on these differences, both on the content and the targeting differences.

```python
ca_advertisers = {}
# 'ca_owner_name': 'Klarna'

def get_targeting(ut, aid):
    # also updates global list of CA advertisers
    global ca_advertisers
    
    # there is always an el['__typename'] == 'WAISTUIAgeGenderType'
    all_targetings = {'id': aid}
    for el in ut['data']['waist_targeting_data']:
        if el['__typename'] == 'WAISTUIInterestsType':
            all_targetings['interests'] = set([i['name'] for i in el['interests']])
        elif el['__typename'] == 'WAISTUICustomAudienceType':
            ca_advertisers[aid] = el['dfca_data']['ca_owner_name']
            all_targetings['custom'] = True
        elif el['__typename'] == 'WAISTUILocationType':
            gran = json.loads(el['serialized_data'])['location_granularity']
            loc = el['location_name']
            all_targetings['location'] = {'loc': loc, 'gran': gran}
        elif el['__typename'] == 'WAISTUIAgeGenderType':
            all_targetings['age-gender'] = {
                'age_min': el['age_min'],
                'age_max': el['age_max'],
                'gender': el['gender']
            }        
        
    return all_targetings
```

```python
from collections import defaultdict

def count_code_props(codes, freqs=None, norm=True):
    # takes dict from adid -> codes and returns dict of code proportions
    counts = defaultdict(lambda: 0)
    for aid in codes:
        # in case of multiple codes, count each one -- essentially computing fraction of codes and not ads here
        for code in codes[aid].split(';'):
            if "Can't determine" not in code and code != 'Study':
                if freqs is not None:
                    # freqs should be index-set dataframe of the participant's ad counts
                    counts[code] += int(freqs.loc[str(aid)])
                else:
                    counts[code] += 1
                
    if norm:
        return {c: counts[c]/sum(counts.values()) for c in counts}
    else:
        return counts
```

## Do Global and Contextual harms focus on different people?

```python
global_harm_counts = {}
cont_harm_counts = {}
healthcare_harm_counts = {}

for pid in pids:
    joiner = pid_adid[pid_adid['pid'] == pid]    
    
    joined = joiner.merge(global_harms, how='inner', left_on='adid', right_on='adid')
    # factor in frequency of each ad
    curr = 0
    for aid in joined['adid']:
        curr += int(adcounts.loc[pid, str(aid)])
    global_harm_counts[pid] = curr
    
    joined = joiner.merge(cont_harms, how='inner', left_on='adid', right_on='adid')
    curr = 0
    for aid in joined['adid']:
        curr += int(adcounts.loc[pid, str(aid)])        
    cont_harm_counts[pid] = curr
    
    joined = joiner.merge(healthcare_harms, how='inner', left_on='adid', right_on='adid')
    curr = 0
    for aid in joined['adid']:
        curr += int(adcounts.loc[pid, str(aid)])
    healthcare_harm_counts[pid] = curr
    
# output these to a file for reusability elsewhere
allrows = []
for pid in pids:
    allrows.append([pid, global_harm_counts[pid], cont_harm_counts[pid], healthcare_harm_counts[pid]])
    
# pd.DataFrame(allrows, columns=['pid', 'global', 'contextual', 'healthcare'])\
#   .to_csv('global_contextual_harmscounts.csv', index=False)
```

#### Statistically Comparing the Distributions

```python
import scipy.stats as stats

# comparing the two distributions shown in the CDF plot; H0 = both distributions are identical
print('=== Global vs. Sensitive ===')
print(stats.mannwhitneyu([global_harm_counts[pid] for pid in part_order],
                        [cont_harm_counts[pid] for pid in part_order]))

# comparing via Kolmogorov-Smirnov test; H0 = both distributions are identical
print(stats.kstest([global_harm_counts[pid] for pid in part_order],
                  [cont_harm_counts[pid] for pid in part_order]))

print('\n=== Global vs. Healthcare Sensitive ===')
print(stats.mannwhitneyu([global_harm_counts[pid] for pid in part_order],
                        [healthcare_harm_counts[pid] for pid in part_order]))
print(stats.kstest([global_harm_counts[pid] for pid in part_order],
                  [healthcare_harm_counts[pid] for pid in part_order]))
```

**Result**: $p < 0.05$ (or 0.025 if corrected here), so $H_0$ (both distributions are equal) for both tests can be rejected.

### Targeting Differences for Top Consumers

```python
# for people who are top consumers of harm (part_names), is the targeting of harmful ads different from 
# their overall targeting? is it more generic than their general ad distribution?

tab = []
types = ['Other', 'P/H/C', 'Sensitive']
for pid in part_names:
    curr_global_harms = list(global_harms.merge(pid_adid[pid_adid['pid'] == pid])['adid'])
    curr_cont_harms = list(cont_harms.merge(pid_adid[pid_adid['pid'] == pid])['adid'])
    curr_all_ads = list(pid_adid[pid_adid['pid'] == pid]['adid'])
    others = list(set(curr_all_ads) - set(curr_global_harms) - set(curr_cont_harms))
        
    for i, adidlist in enumerate([others, curr_global_harms, curr_cont_harms]):
        total = len(adidlist)
        n_interest, n_custom, n_generic = 0, 0, 0        
        for aid in adidlist:
            # preprocess targeting json
            tar = get_targeting(ad_targetings[str(aid)], aid)
            if 'interests' in tar:
                n_interest += 1
            if 'custom' in tar:
                n_custom += 1
            
        n_generic = total - n_interest - n_custom        
        tab.append([part_names[pid], types[i], total, n_generic, n_interest, n_custom])
    tab.append([])
    
print(tabulate(tab, headers=['Participant', 'Type', 'Total', 'Generic', 'Interest', 'Custom']))
```

**NOTE**: I'm onto something here -- the globally harmful ads have a higher fraction of generically targeted ads than benign (~72% vs. 48.6%), this is where FB is doing most of the work. For contextual harms, custom targeting makes up 31% of the targeting, compared to 25% for benign ones; 27% sensitive are targeted with interest compared to 25% benign. I think these are almost fundamental insights about the algorithm vs. advertiser equation.
